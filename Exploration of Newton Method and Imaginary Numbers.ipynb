<div> What is Newton's Method?</div>
<div>Newton's Method is an iterative method of finding a convergence of a root.</div>
<br>
<div> A convergence of a root is when the change in the root has become so minimal to the point where it is equal. We will determine the minimal as when point $x_n$ is within 15 decimal places of $x_{n+1}$</div>

The Newton's Method is defined as $$x_{n+1} = x_n - {\frac{f(x_n)} {f'(x_n)}}$$

<div>
Given an initial guess $x_n$, we will find $x_{n+1}$ using Newton's Method until the initial point converges
</div>
<br>
<div>In order to find the root for convergence, we need to use Euler's Formula. 
    <div>Euler's Formula is defined as:</div>
$$\frac{cos(2 \pi k)}{n} + i\frac{sin(2 \pi k)}{n}$$ where k is an array [0,1..n-1] 
</div>
<br>
<div>
Euler's Formula will somtimes return complex roots. Complex roots are defined as imaginary numbers. Imaginary numbers happen when you take a square root of a negative number.

<h3>Part One</h3>
Let us Explore the function $x^3 - 7x^2 + 12x$

```python
#imports
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x is a number, f, df are functions and n is the power of x
def newton(x,f,df,n):
  #if the divisor is not 0, then return the answer
  if(df(x,n) !=0):
      #ans is a complex number/normal of the result of newton
      ans =  complex(x - f(x,n)/df(x,n))
      return ans
  return 0
```

<h4>Define: newton(x,f,df,n):</h4>
newton takes in a number, X, and 2 functions. f represents the normal functions and df represents the derivative of the function. n is the power of x. The function outputs the next guess using the formula (x - f(x,n)/df(x,n))

```python
#x is a number, n is the power of x
def fun1(x,n):
    #returns the result of x^3 - 7x^2 + 12x
  return math.pow(x.real,3) - 7*math.pow(x.real,2) + 12*x.real
def dfun1(x,n):
    #returns the result of 3x^2 - 14x + 12
  return 3*math.pow(x.real,2) - 14*x.real + 12
```

<h4>Define fun1(x,n), dfun1(x,n):</h4>
fun1 and dfun1 takes in a number and a power and returns the result of the function $x^3 - 7x^2 + 12x$ and $3x^2 - 14x + 12$

```python
#num is the number we can to perform newton on, f, df are functions and n is the power of num
def pointsOfConvergence(num,f,df,n):
  #holds the current value of num
  holder = 0
  #counts the amount of iterations until changes are minimal
  counter = 0
  for i in range (1000):
    #Round num to 15 decimal places
    num = "{:.15f}".format(num)
    #Turn num into a complex number
    num = complex(num)
    #Assign the result of newton to holder
    holder = newton(num,f,df,n)
    #Round the result to 15 decimal places and turn it into a complex number
    holder = "{:.15f}".format(holder)
    holder = complex(holder)
    #check if previous iteration is equal to current iteration, if it is then stop otherwise continue
    if(holder !=num):
      counter +=1
      num = holder
    else:
      break
 #num is the number it converges to and counter is the amount of iteration it needs
  return [num,counter]
```

<h4>Define pointsOfConvergence(num,f,df,n): </h4>
pointsOfConvergence takes in a number n, two functions f,df and n. It uses these parameters to call newton and find the number that it converges to as well as the amount of iterations it takes. It returns the converged number and the amount of iterations

```python
#start is the start point of the range, end is the end point of the range and amt is the number of points inbetween
def part1Graph(start,end,amt):
  #create an array using start,end,amt
  initGuess = np.linspace(start,end,amt)
  #find the root the points in initGuess converges to using pointsOfConvergence
  roots = [pointsOfConvergence(point,fun1,dfun1,3)[0].real for point in initGuess]
  #Graph the results
  plt.title("Graph for $x^3 - 7x^2+12x$")
  plt.xlabel("Initial Guess$_0$")
  plt.ylabel("Roots")
  plt.plot(initGuess,roots, 'o');
```

<h4>Define part1Graph(start,end,amt): </h4>
part1Graph takes in a start range start, an end range end, and the an amount of points in between amt. It uses this to generate a list of points to find a root it converges to and plots it

<h3>Part Two</h3>
Let us explore the imaginary numbers. The function we will look at for part two is $z = z^3-1$

```python
#this has the same concept as fun1, and dfun1 except with a different equation
def func2(z,n):
  #return z^n -1
  return pow(z,n)-1
def dfunc2(z,n):
  #return the derivative of z^n-1
  return n*pow(z,(n-1))
```

<h4>Define func2(z,n), dfunc2(z,n): </h4>
func2, and dfunc2 computes the function $z=z^n-1$ and the derivative of the function $z= 2z^{n-1}$ where z is the number and n is the power we want to compute it at.

```python
def nroot(n):
  #use Euler's Formula to compute the roots of n and return it
  return [complex(np.cos(2*np.pi*k/n),np.sin(2*np.pi*k/n)) for k in range(n)]
```

<h4>Define nroot(n):</h4>
nroot takes in a number n and uses Euler's formula to compute an array of roots for the given number and returns it

```python
#example of nroot 3
nroot(3)
```

```python
#num is a number and listOfRoots is an array of roots
def imagineNumber(num,listOfRoots):
  #converts num and the listOfRoots to round to a decimal place of 2
  num = complex("{:.2f}".format(num))
  for i in range(len(listOfRoots)):
    comp = complex("{:.2f}".format(listOfRoots[i]))
    #if num and comp are the same, return i, which is the index of the root.
    #We will use this to determine different roots
    if(num == comp):
      return i
  return -1
```

<h4>Define imagineNumber(num, listOfRoots)</h4>

```python
def plotter(point,val):
  colors = ["red","lime","blue","orange","purple","pink","green","cyan","brown","olive","gray","hotpink"]
  plt.scatter(point.real, point.imag, marker='o', color = colors[val])
```

<h4>Define plotter(point,val)</h4>

```python
#start is the start point of the range, end is the end point of the range and amt is the points in between
def allpointsWithin(start,end,amt):
  #create an array with the range start, end with amt amount of points
  arr = np.linspace(start,end,amt)
  #turns the array into an array of imaginary numbers
  imagineNum = [p*1j for p in arr]
  #create a list of all combinations of arr points with its respective imaginary points
  allPoints = [xp+yp for yp in imagineNum for xp in arr]
  return allPoints
```

<h4>Define allpointsWithin(start,end,amt): </h4>
allpointsWithin takes in a start and end range with amt amount of points. It then creates an array with all the possibilites of arr with arr amount of imaginary numbers and returns it. This is essentially a 2d matrix with real and imaginary numbers.

```python
#start, end are the range, amt is the point in between, f, df are functions, and color is the color map
def part2(start,end,amt,f,df,color):
  #get the root for n=3
  threeRoot = nroot(3)
  #generate all possible combinations with start, end, amt
  allPoints = allpointsWithin(start,end,amt)
  #use point of convergence to see what point the number converges to and use imaginenumber to determine what root it
  #is equal to and put it all into conRoots
  conRoots = [imagineNumber(pointsOfConvergence(point,f,df,3)[0],threeRoot ) for point in allPoints]
  plt.title("Graph for $Z = Z^3-1$")
  plt.xlabel("Real Numbers")
  plt.ylabel("Imaginary Numbers")
  #reshape the array into a 2d matrix
  x = np.reshape(conRoots, (amt, amt))
  #plot the points with the given color map
  plt.imshow(x ,cmap=color);
```

<h4>Define part2(start,end,amt,f,df,color):</h4>

Let us explore how the amount of iterations on a point and then plot each different amount of iteration as a different color.

```python
#n is the power of z, start, end are the range, amt is the amount of points, 
#f,df are the function and derivative function, color is the color map you want to use
def graphBrightness(n,start,end,amt,f,df,colorMap):
  #get the roots to the power of n
  roots = nroot(n)
  #create an array with all points within amt and amtj
  allPoints = allpointsWithin(start,end,amt)
  #use point of convergence to see what point the number converges to and 
  #use imaginenumber to determine the amount of iterations for the convergence
  conRoots = [pointsOfConvergence(point,f,df,n)[1] for point in allPoints]
  plt.title("Brightness for $Z = Z^{"+str(n)+"}-1$")
  plt.xlabel("Real Numbers")
  plt.ylabel("Imaginary")
  #turn the arr of points into a matrix of amt x amt size
  x = np.reshape(conRoots, (amt,amt))
  plt.imshow(x ,cmap=colorMap);
  return conRoots
```
